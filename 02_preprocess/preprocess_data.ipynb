Null values

```python
def missing_values_summary(df):
    total_rows = df.shape[0] # total rows
    null_counts = df.isnull().sum() # nulls for each column
    null_percentages = (null_counts / total_rows) * 100 # percentage of nulls for each feature
    
    # DataFrame to store the results
    missing_summary = pd.DataFrame({
        'Feature': df.columns,
        'Total Rows': total_rows,
        'Null Count': null_counts,
        'Null Percentage': null_percentages
    })
    
    # Sort DataFrame by the percentage nulls 
    missing_summary = missing_summary.sort_values(by='Null Percentage', ascending=False).reset_index(drop=True)    
    return missing_summary
```

```python
# Example usage:
train_missing_summary = missing_values_summary(train_df)
test_missing_summary = missing_values_summary(test_df)
```

### Preprocess Data

Note: 
I will attempt 2 model at inference. XGBoost and Catboost

- XGBoost is more suited to datasets with higher number of numerical features
- Catboost is more suited to datasets with higher number of categorical features

The preprocessing required for both models differs in the handling of categorical features. Catboost requires that we do NOT create dummy features. 

Hence I will create a train, test and oot DataFrame for each model. 

##### Preprocess Strategy: 
- Drop extremely high null features. (missing > 75% of data in train set)
- Define numerical, categorical and date time features and convert to appropriate dtype.
- Define irrelevant features and drop them from the data.
- Impute missing values. (median value for numerical for skewness resistance and unkown value for categorical features)
- Calculate time deltas for datetime features.
- Drop unnecessary features

_Only applicable to XGB ready data_
- Determine which categorical features to use in model and convert to dummies. 
- Create dummy features for categorical features.
- Align dummy features in train and test.
